```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
acctrain = pd.read_csv('acctrain.csv')
accval = pd.read_csv('accval.csv')
losstrain = pd.read_csv('losstrain.csv')
lossval = pd.read_csv('lossval.csv')
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(6,6))

sns.lineplot(data=acctrain, x='Step', y='Value')
plt.title('Accuracy during training for each epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy Ratio')
plt.show()
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(6,6))

sns.lineplot(data=accval, x='Step', y='Value')
plt.title('Accuracy during validation for each epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy Ratio')
plt.show()
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(6,6))

sns.lineplot(data=losstrain, x='Step', y='Value')
plt.title('Loss during training for each epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(6,6))

sns.lineplot(data=lossval, x='Step', y='Value')
plt.title('Loss during validation for each epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
